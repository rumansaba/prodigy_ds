```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python

# Load the Titanic dataset
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)
```

```python
df
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['Age'].fillna(df['Age'].mean(), inplace=True)
```

```python
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
```

```python
# too many values missing so we delete the column cabin

df.drop(columns=['Cabin'], inplace=True)
```

```python
df.isnull().sum()
```

```python
# Convert 'Sex' and 'Embarked' to numerical values
df['Sex'] = df['Sex'].map({'male': 0, 'female': 1})
df['Embarked'] = df['Embarked'].map({'C': 0, 'Q': 1, 'S': 2})
```

```python
df['Fare']
```

```python
df['Embarked']
```

```python
df['Survived']
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set up the visualizations
sns.set(style='whitegrid')

# Plot survival rate
sns.countplot(x='Survived', data=df)
plt.title('Survival Count')
plt.show()

```

```python
# Plot survival rate by gender
sns.countplot(x='Survived', hue='Sex', data=df)
plt.title('Survival Count by Gender')
plt.show()

```

```python
survived = df['Survived'].sum()
not_survived = (df['Survived'] == 0).sum()
print(survived)

print(not_survived)

print(df['Sex'].unique())
print(df['Survived'].unique())

# Count the number of females who did not survive
female_not_survived_count = df[(df['Sex'] == 1) & (df['Survived'] == 0)].shape[0]
print(f'Number of female passengers who did not survive: {female_not_survived_count}')

female_survived_count = df[(df['Sex'] == 1) & (df['Survived'] == 1)].shape[0]
print(f'Number of female passengers who did not survive: {female_survived_count}')

male_not_survived_count = df[(df['Sex'] == 0) & (df['Survived'] == 0)].shape[0]
print(f'Number of female passengers who did not survive: {male_not_survived_count}')

male_survived_count = df[(df['Sex'] == 0) & (df['Survived'] == 1)].shape[0]
print(f'Number of female passengers who did not survive: {male_survived_count}')
```

```python
# Plot survival rate by class
sns.countplot(x='Survived', hue='Pclass', data=df)
plt.title('Survival Count by Class')
plt.show()

```

```python
# Plot age distribution
sns.histplot(df['Age'], bins=30, kde=True)
plt.title('Age Distribution')
plt.show()

```

```python
# Plot fare distribution
sns.histplot(df['Fare'], bins=30, kde=True)
plt.title('Fare Distribution')
plt.show()

```

```python
# Plot age vs. fare
sns.scatterplot(x='Age', y='Fare', hue='Survived', data=df)
plt.title('Age vs. Fare by Survival')
plt.show()
```

```python
# Survival rate by age
sns.histplot(data=df, x='Age', hue='Survived', multiple='stack', bins=30)
plt.title('Survival Rate by Age')
plt.show()

```

```python
# Survival rate by fare
sns.histplot(data=df, x='Fare', hue='Survived', multiple='stack', bins=30)
plt.title('Survival Rate by Fare')
plt.show()

```

```python
# Survival rate by embarked location
sns.barplot(x='Embarked', y='Survived', data=df)
plt.title('Survival Rate by Embarked Location')
plt.show()
```
