```python
gender_count = data['Region'].value_counts()
bar_width = 0.9
x = range(len(gender_count.index))
```

```python
plt.bar(gender_count.index, gender_count.values)
plt.xlabel('Region')
plt.ylabel('Count')
plt.title('Distribution of Region')

plt.xticks(x, gender_count.index, rotation=45)
plt.show()
```
